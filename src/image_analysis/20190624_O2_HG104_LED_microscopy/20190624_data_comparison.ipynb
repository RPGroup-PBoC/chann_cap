## Comparing the data

For this dataset taken on `20190624` I used a Nikon scope that had an LED illuminaiton system rather than a single wavelength laser line. The reason I took this dataset was to compare the cell-to-cell variability in the fluorescence values with both illumination systems.

These systems are different in the following aspects:
- With the LED system rather than having a single wavelength exciting the fluorophores there is a larger range of wavelengths that go through.
- The spatial structure of the LED system is much uniform compared to the laser. That is why the data didn't even need to be flatten because the illumination profile is practially flat.

So let's read both data sets into memory. First the LED data

```python
df_led = pd.read_csv('outdir/20190624_O2_HG104_IPTG_titration_microscopy.csv',
                     comment='#')

df_led[['date', 'operator', 'rbs', 'mean_intensity', 'intensity']].head()
```

Now the rest of the datasets taken with the laser system

```python
# Read the tidy-data frame
files = glob.glob(datadir + '/*IPTG*csv')# + mwc_files
df_laser = pd.concat(pd.read_csv(f, comment='#') for f in files if 'Oid' not in f)

##  Remove data sets that are ignored because of problems with the data quality
##  NOTE: These data sets are kept in the repository for transparency, but they
##  failed at one of our quality criteria
##  (see README.txt file in microscopy folder)
ignore_files = [x for x in os.listdir('../../image_analysis/ignore_datasets/')
                if 'microscopy' in x]
# Extract data from these files
ignore_dates = [int(x.split('_')[0]) for x in ignore_files]

# Remove these dates
df_laser = df_laser[~df_laser['date'].isin(ignore_dates)]

# Keep only the O2 operator
df_laser = df_laser[df_laser.operator == 'O2']

df_laser[['date', 'operator', 'rbs', 'mean_intensity', 'intensity']].head()
```

Let's now look at the O2 $\Delta lacI$ strain data. For this we first have to extract the mean autofluorescence value. First let's process the LED data.

```python
# Define names for columns in dataframe
names = ['date', 'IPTG_uM','operator', 'binding_energy',
        'rbs', 'repressors', 'mean', 'std', 'noise']

# Initialize df_led frame to save the noise
df_noise_led = pd.DataFrame(columns=names)

# Extract the mean autofluorescence
I_auto = df_led[df_led.rbs == 'auto'].intensity.mean()

# Extract the strain fluorescence measurements
strain_df_led = df_led[df_led.rbs == 'delta']

# Group df_led by IPTG measurement
df_led_group = strain_df_led.groupby('IPTG_uM')
for inducer, df_led_inducer in df_led_group:
    # Append the require info
    strain_info = [20190624, 0, df_led_inducer.operator.unique()[0],
                   df_led_inducer.binding_energy.unique()[0],
                   df_led_inducer.rbs.unique()[0],
                   df_led_inducer.repressors.unique()[0],
                  (df_led_inducer.intensity - I_auto).mean(), 
                  (df_led_inducer.intensity - I_auto).std(ddof=1)]
    # Check if the values are negative for very small noise
    if strain_info[int(np.where(np.array(names) == 'mean')[0])] > 0:
        # Compute the noise
        strain_info.append(strain_info[-1] / strain_info[-2])
        # Convert to a pandas series to attach to the df_ledframe
        strain_info = pd.Series(strain_info, index=names)
        # Append to the info to the df_led frame
        df_noise_led = df_noise_led.append(strain_info, 
                                               ignore_index=True)

df_noise_led.head()
```

```python
# group by date and by IPTG concentration
df_group = df_laser.groupby(['date'])

# Define names for columns in data frame
names = ['date', 'IPTG_uM','operator', 'binding_energy',
        'rbs', 'repressors', 'mean', 'std', 'noise']

# Initialize data frame to save the noise
df_noise_delta = pd.DataFrame(columns=names)
for date, data in df_group:
    # Extract the mean autofluorescence
    I_auto = data[data.rbs == 'auto'].intensity.mean()
    
    # Extract the strain fluorescence measurements
    strain_data = data[data.rbs == 'delta']
    
    # Group data by IPTG measurement
    data_group = strain_data.groupby('IPTG_uM')
    for inducer, data_inducer in data_group:
        # Append the require info
        strain_info = [date, inducer, data_inducer.operator.unique()[0],
                       data_inducer.binding_energy.unique()[0],
                       data_inducer.rbs.unique()[0],
                       data_inducer.repressors.unique()[0],
                      (data_inducer.intensity - I_auto).mean(), 
                      (data_inducer.intensity - I_auto).std(ddof=1)]
        # Check if the values are negative for very small noise
        if strain_info[int(np.where(np.array(names) == 'mean')[0])] > 0:
            # Compute the noise
            strain_info.append(strain_info[-1] / strain_info[-2])
            # Convert to a pandas series to attach to the dataframe
            strain_info = pd.Series(strain_info, index=names)
            # Append to the info to the data frame
            df_noise_delta = df_noise_delta.append(strain_info, 
                                                   ignore_index=True)
    
df_noise_delta.head()
```

It seems that the noise is exactly the same for both illumination systems, â‰ˆ 0.4-0.5.

Let's look at the ECDF of single-cell fluorescence values. For all measurements to be comparable we will plot the fold-change distribution. What this means is that we will extract the mean autofluorescence value and we will normalize by the mean intensity of the $\Delta lacI$ strain.

```python
# group laser data by date
df_group = df_laser.groupby('date')

colors = sns.color_palette('Blues', n_colors=len(df_group))
# Loop through dates
for j, (g, d) in enumerate(df_group):
    # Extract mean autofluorescence
    auto = d.loc[d.rbs == 'auto', 'intensity'].mean()
    # Extract mean delta
    delta = d.loc[d.rbs == 'delta', 'intensity'].mean()
    # Keep only delta data
    data = d[d.rbs == 'delta']
    fold_change = (data.intensity - auto) / (delta - auto)
    # Generate ECDF
    x, y = ccutils.stats.ecdf(fold_change)
    # Plot ECDF
    plt.plot(x[::10], y[::10], lw=0, marker='.', color=colors[j],
               alpha=0.3, label='')

## LED
# Extract mean autofluorescence
auto_led = df_led.loc[df_led.rbs == 'auto', 'intensity'].mean()
delta_led = df_led.loc[df_led.rbs == 'delta', 'intensity'].mean()

fold_change = (df_led[df_led.rbs == 'delta'].intensity - auto_led) /\
              (delta_led - auto_led)   
# Generate ECDF
x, y = ccutils.stats.ecdf(fold_change)
# Plot ECDF
plt.plot(x, y, lw=0, marker='v', color='red',
           alpha=0.3, label='LED', ms=3)

# Add fake plot for legend
plt.plot([], [], marker='.', color=colors[-1],
         alpha=0.3, label='laser', lw=0)

# Label x axis
plt.xlabel('fold-change')

# Add legend
plt.legend()

# Label y axis of left plot
plt.ylabel('ECDF')

# Change limit
plt.xlim(right=3)

plt.savefig('outdir/ecdf_comparison.png', bbox_inches='tight')
```

The data seems to completely overlap with the previous datasets. So the spatial structure of the LED camera did not change the single-cell distribution.
